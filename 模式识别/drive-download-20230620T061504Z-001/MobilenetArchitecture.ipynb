```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2

# 创建一个预训练的 MobileNetV2 模型
model = MobileNetV2(weights='imagenet')

```

```python
if isinstance(model, tf.keras.Sequential):
    print("这是一个 Sequential 模型")
elif isinstance(model, tf.keras.Model):
    print("这是一个函数式 API 模型")
else:
    print("这不是一个 Keras 模型")
```

```python
model.summary()
```

```python
# 获取模型的输入和倒数第二层的输出
input_tensor = model.input
output_tensor = model.layers[-2].output

# 创建一个新的模型
new_model = tf.keras.Model(inputs=input_tensor, outputs=output_tensor)
new_model.summary()  # 显示更新后的模型结构
```

```python
# 计算模型中层数量的一半
num_layers = len(new_model.layers)
half_layers = num_layers // 2

# 冻结前半部分的层
for layer in new_model.layers[:half_layers]:
    layer.trainable = False

# 显示每一层的 trainable 属性
for layer in new_model.layers:
    print(f"{layer.name}: {layer.trainable}")
```
