```python
import re

s = "3244 (_sex male) (_age senior) (_race white) (_face serious) (_prop '(moustache beard ))"
result = re.sub(r"((_sex (.?) )) ((_age (.?) )) ((_race (.?) )) ((_face (.?) )) ((_prop '(.*?)' ))", r'\2,\4,\6,\8,\10,', s)
result = result.strip(',')  # 去掉最后一个逗号
print(result)
```

```python
# GPT4
# import re

input_string = "3244 (_sex male) (_age senior) (_race white) (_face serious) (_prop '(moustache beard ))"
pattern = r"\(_\w+ '(.*?) ?\)|\(_\w+ (.*?) ?\)"
replacement = r"\1\2,"

output_string = re.sub(pattern, replacement, input_string).rstrip(',')
print(output_string)
```

```python
# import re

string = "3244 (_sex male) (_age senior) (_race white) (_face serious) (_prop '(moustache beard ))"
pattern = r"\s*\(_\w+\s+([^)]+)\)"
replacement = r",\1"
result = re.sub(pattern, replacement, string)
result = result.replace("'", "")
result = result.replace("(", "")
result = result.replace(")", "")
print(result)
```
